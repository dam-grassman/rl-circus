## <a id="Jacobian-Saliency-Map"></a>Jacobian Saliency Map

**Paper** : *Dueling Network Architectures for Deep Reinforcement Learning*, Whang et al 2016

In the paper that introduces the Dueling Network Architectures for DQN, the authors added a short section on saliency maps visualization using a gradient-based method (see *Simonyan et al., 2013*).	
They do so by computing the **absolute value of the Jacobian** of the neural network w.r.t the input frames $s$ : $\nabla_s \hat{V}(s,\theta)$.

As a reminder, the Jacobian matrix of $f : \mathbb{R}^n \longleftarrow \mathbb{R}^n$ wrt to its input $x \in \mathbb{R}^n$ is a $m \times n$ matrix, denoted **J** (or $\nabla f$ whose $(i,j)^{th}$ entry is $J_{ij} = \frac{\partial f_i}{\partial x_j}$, or explicitely :

$$ J = [\frac{\partial f}{\partial x_1}, ... , \frac{\partial f}{\partial x_n}] = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & ... & \frac{\partial f_1}{\partial x_n}\\
... &  & ...\\
\frac{\partial f_m}{\partial x_1} & ... & \frac{\partial f_m}{\partial x_n}
\end{bmatrix}$$

For an image as input $x \in \mathbb{R}^{w \times h \times c}$, the Jacobian matrix of of a classifier can also be represented as an image, or an heatmap, of the partial deriviative w.r.t each pixel.

We will be interesting in computing the Jacobian matrix w.r.t the two different outputs : on the one hand, **actor** part (pi) that returns a logits of which action to select and on the other hand, the **critic** part (v) that evaluate the value of the state.

```python
def get_jacobian(model, layer, output, input):
    ''' Inspire from github.com/greydanus/visualize_atari'''
    global top_h_ ; top_h_ = None
    def hook_top_h(m, i, o): global top_h_ ; top_h_ = o.clone()
    hook1 = layer.register_forward_hook(hook_top_h)
    _ = model(input)
    print(top_h_, output)
    torch.autograd.backward(top_h_, output, retain_graph = True)
    hook1.remove()
    return input.grad.data.clone()
```

```python
state = Variable(history['obs'][idx].cuda(),  requires_grad=True)
jacb = get_jacobian(policy, policy.v, history['values'][idx].cuda(), state)
```

```python
from skimage.transform import resize
def saliency_on_atari_frame(saliency, atari, fudge_factor, channel=2, sigma=0, size = [84,84]):
    pmax = saliency.max()
    #S = imresize(saliency, size=size, interp='bilinear').astype(np.float32)
    S = resize(saliency, output_shape=size, mode='reflect').astype(np.float32)
    S = S if sigma == 0 else gaussian_filter(S, sigma=sigma)
    S -= S.min() ; S = fudge_factor*pmax * S / S.max()
    I = atari.astype('uint16')
    I[:,:,channel] += S.astype('uint16')
    I = I.clip(1,255).astype('uint8')
    return I
```

```python
fig, ax = plt.subplots(1,3, figsize=(10,5))
for a in fig.axes: a.get_xaxis().set_visible(False) ; a.get_yaxis().set_visible(False) ;

frame = (np.expand_dims(history['obs'][idx].squeeze().numpy()[3,:,:],-1)*255).astype('uint8')
frame = np.concatenate([frame, frame, frame], axis=-1)
jacobian = jacb.cpu().numpy()[0,3,:,:]
saliency = saliency_on_atari_frame((jacobian**2).squeeze(), frame, fudge_factor=50, channel=2, sigma=0)

ax[0].imshow(frame) ; ax[0].set_title('Original Image')
ax[1].imshow(jacobian) ;  ax[1].set_title('Jacobian Saliency Map')
ax[2].imshow(saliency) ;  ax[2].set_title('Final Render')
plt.show()
```

```python
from __future__ import print_function
import warnings ; warnings.filterwarnings('ignore') # mute warnings, live dangerously

import matplotlib.pyplot as plt
import matplotlib as mpl ; mpl.use("Agg")
import matplotlib.animation as manimation
import matplotlib.patches as patches

import torch
from torch.autograd import Variable
import torch.nn.functional as F

import gym, os, sys, time, argparse
sys.path.append('C:/Users/damie/Desktop/XRL/Code/SALIENCYMAP')
from visualize_atari import *
```

```python
env_name = 'Breakout-v0'
save_dir = 'figures/'

import numpy as np
np.random.seed(1)

print("set up dir variables and environment...")
load_dir = 'C:/Users/damie/Desktop/XRL/Code/SALIENCYMAP/visualize_atari/{}/'.format(env_name.lower())
meta = get_env_meta(env_name)
env = gym.make(env_name) ; env.seed(1)

%matplotlib inline

print("initialize agent and try to load saved weights...")
model = NNPolicy(channels=1, num_actions=env.action_space.n)
_ = model.try_load(load_dir, checkpoint='*.tar') ; torch.manual_seed(1)

print("get a rollout of the policy...")
history = rollout(model, env, max_ep_len=3e3)
```

```python
frame_ix= 1890
def jacobian(model, layer, top_dh, X):
    global top_h_ ; top_h_ = None
    def hook_top_h(m, i, o): global top_h_ ; top_h_ = o.clone()
    hook1 = layer.register_forward_hook(hook_top_h)
    _ = model(X) # do a forward pass so the forward hooks can be called

    # backprop positive signal
    #print(top_h_.shape, top_dh.shape)
    torch.autograd.backward(top_h_, top_dh.clone(), retain_graph = True) # backward hooks are called here
    hook1.remove()
    return X[0].grad.data.clone().numpy(), X[0].data.clone().numpy()

# derivative is simply the output policy distribution
top_dh_actor = torch.Tensor(history['logits'][frame_ix]).view(1,-1)
top_dh_critic = torch.Tensor(history['values'][frame_ix]).view(1,-1).fill_(1)

# get input
tens_state = torch.Tensor(prepro(history['ins'][frame_ix]))
state = Variable(tens_state.unsqueeze(0), requires_grad=True)
hx = Variable(torch.Tensor(history['hx'][frame_ix-1]).view(1,-1))
cx = Variable(torch.Tensor(history['cx'][frame_ix-1]).view(1,-1))
X = (state, (hx, cx))
actor_jacobian, _ = jacobian(model, model.actor_linear, top_dh_actor, X)
state.grad.mul_(0) ; X = (state, (hx, cx))
critic_jacobian, _ = jacobian(model, model.critic_linear, top_dh_critic, X)
```
